# **TITLE : Mileage Prediction**

# **Objective :**

The objective of a mileage prediction Python model is to estimate or predict the fuel efficiency (mileage) of a vehicle, typically measured in miles per gallon (MPG) or kilometers per liter (km/L), based on various input features. 
# **Source :**

The data set consists of 2000 samples from each of two categories. 
Four variables are
1. Mpg	
2. Cylinders	
3. Displacement	
4. Horsepower	
5. Weight	
6. Acceleration	
7. Model year	
8. Origin	
9. Name


```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/refs/heads/main/MPG.csv')
```

```python
df.head()
```

```python
df.nunique()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
sns.pairplot(df, x_vars=['displacement', 'horsepower', 'weight', 'acceleration', 'mpg'], y_vars=['mpg']);
```

```python
sns.regplot(x='displacement', y='mpg', data=df);
```

```python
df.columns
```

```python
y = df['mpg']
```

```python
y.shape
```

```python
X = df[['displacement', 'horsepower', 'weight', 'acceleration']]
```

```python
X.shape
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

```python
pd.DataFrame(X).describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
```

```python
X_train2 = poly.fit_transform(X_train)
```

```python
X_test2 = poly.fit_transform(X_test)
```

```python
lr.fit(X_train2, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly = lr.predict(X_test2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred_poly)
```

```python
mean_absolute_percentage_error(y_test, y_pred_poly)
```

```python
r2_score(y_test, y_pred_poly)
```
